```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('diabetes_prediction_dataset.csv')
data.head()
```

```python
print(data.shape)
print(data.isnull().sum())
```

```python
count_0, count_1 =data['diabetes'].value_counts()
```

```python
data_0 = data.loc[data['diabetes'] == 0]
data_1 = data.loc[data['diabetes'] ==1]
print(data_0.head())
print(data_1.head())
```

```python
#OVER SAMPLING DATA

data_1_over = data_1.sample(count_0, replace=True)
```

```python
data_1_over.head(10)
```

```python
df = pd.concat([data_0, data_1_over], axis=0,ignore_index=True)
df.shape
```

```python
df.tail(10)
```

```python
df.diabetes.value_counts()
```

```python
print(df['smoking_history'].value_counts())
print(df['gender'].value_counts())
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder

label_gender = LabelEncoder()
ohe_smoking = OneHotEncoder()
standard_scaler = StandardScaler()
```

```python
df['gender'] = np.array(label_gender.fit_transform(df['gender']))
df.head()
```

```python
smoke_encoder = ohe_smoking.fit_transform(df[['smoking_history']])
print(ohe_smoking.get_feature_names_out(['smoking_history']))
```

```python
smoke_encoder.toarray()
```

```python
smoking_df = pd.DataFrame(smoke_encoder.toarray(), columns=ohe_smoking.get_feature_names_out(['smoking_history']))
smoking_df.head()
```

```python
smoking_df = smoking_df.reindex(index= range(len(df)))
df=df.reindex(index= range(len(df)))
```

```python
df = pd.concat([df.drop('smoking_history', axis =1),smoking_df], axis=1)
```

```python
df['diabetes'].value_counts()
```

```python
x = df.drop('diabetes', axis=1)
y = df['diabetes']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=10, stratify=y)
```

```python
x_train
```

```python
x_train = standard_scaler.fit_transform(x_train)
x_test = standard_scaler.transform(x_test)
```

```python
import pickle

with open('scaler.pkl', 'wb') as f:
    pickle.dump(standard_scaler,f)
with open('gender_encoder.pkl', 'wb') as f:
    pickle.dump(label_gender,f)
with open('smoking_encoder.pkl', 'wb') as f:
    pickle.dump(ohe_smoking,f)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
```

```python
x_train.shape
```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
##setting the tensorboard
import datetime
log_dir = 'logs/fit/' + datetime.datetime.now().strftime('%Y-%m-%d')
tensorflow_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
# early stopping

early_stopping_callback = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
#model training

history = model.fit(
                    x_train,y_train,validation_split=0.2,
                    epochs=10,
                    callbacks=[tensorflow_callback, early_stopping_callback]
                    )
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_final = [1 if i>0.5 else 0 for i in y_pred]
y_final
```

```python
model.save('diabetes_model.h5')
```
